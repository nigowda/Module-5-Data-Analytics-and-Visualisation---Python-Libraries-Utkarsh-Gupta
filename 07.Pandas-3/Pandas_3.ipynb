```python
data = movies.merge(directors, how = 'left',left_on = 'director_id', right_on = 'id' )
```

```python
data.drop(['director_id','id_y'],axis=1, inplace = True)
```

```python
def encode(data):
  if data == "Male":
    return 0
  else:
    return 1
```

```python
data['gender'] = data['gender'].apply(encode)
```

```python
# Let's say, we want to filter the risky movies whose budget was even higher than the average revenue of the director from all his  other movies.
```

```python
def func(x):
  x['risky'] = x['budget'] - x['revenue'].mean() >= 0
  return x
```

```python
data.groupby("director_name").apply(func)
```

```python
data.groupby("director_name", group_keys= True).apply(func).reset_index(drop = True)
```
